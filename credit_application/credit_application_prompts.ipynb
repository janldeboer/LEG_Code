```python
def format_application_for_prompt(application_data: pd.Series, credit_application_template: str) -> str:
	return credit_application_template.format(**application_data)

def format_prompt(application_form: str, llm_prompt_template: str) -> str:
	return llm_prompt_template.format(application_form=application_form)
```

```python
df = pd.read_csv(CREDIT_APPLICATANT_DATA_FILE)
```
